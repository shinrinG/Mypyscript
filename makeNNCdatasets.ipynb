```python
import os
import sys
import shutil
import glob
from PIL import Image,ImageOps

data=glob.glob('C:\\Users\\isc\\Desktop\\201910101303_d3qa\\99_temp\\ALLINVED\\*')
count=0
for i in data:
    im=Image.open(i)
    nameseed=os.path.basename(i)
    for n in range(217):
        count=count+1
        left=16*n
        upper=100
        right=16*n+128
        lower=228
        im_crop=im.crop((left,upper,right,lower))

        numbering="{0:03d}".format(n)
        savepath="C:\\Users\\isc\\Desktop\\201910101303_d3qa\\99_temp\\padd\\"+ numbering +'_'+ nameseed
        im_crop.save(savepath)
        out=str(count)+"/434,000"
        print(out)
```
