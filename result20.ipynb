```python
# display a 2D manifold of the digits
n = 21  # figure with 15x15 digits
digit_size = 128
figure = np.zeros((digit_size * n, digit_size * n,3))

import pandas as pd
csvpath="C:\\Users\\isc\\Desktop\\201910101303_d3qa\\04_generate_20_prd\\output_result.csv"
df = pd.read_csv(csvpath)
lists = df.values.tolist()
lists = np.array(lists)
pred = lists[:,2]
pred = pred.tolist()

path = "C:\\Users\\isc\\Desktop\\201910101303_d3qa\\04_generate_20\\"

for i in range(n):
    for j in range(n):
        num=j*21+i
        target = path + str(num) + ".png"
        img = cv2.imread(target)
        chg= img[:,:,0].reshape(128,128,1)
        cha= np.zeros((128,128,1)).reshape(128,128,1)
        chb= np.zeros((128,128,1)).reshape(128,128,1)


        if pred[num]<0.9:
            img2=np.concatenate((chb,cha,chg),axis=2)
            img3=np.int_(img2)
        else:
            img2=np.concatenate((chg,cha,chb),axis=2)
            img3=np.int_(img2)
        
        figure[j * digit_size: (j + 1) * digit_size,
              i * digit_size: (i + 1) * digit_size] = img3
        figure=np.int_(figure)

plt.figure(figsize=(10, 10))
plt.imshow(figure)
plt.show()
```

```python
cv2.imwrite("2Dmani20_09.bmp",figure)
```
